### Creating random tensors
Random tensors are tensors of some arbitrary size with random numbers


```python
random_1 = tf.random.Generator.from_seed(42) #set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2  = random_2.normal(shape=(3,2))

#Are they equal
random_1, random_2, random_1 == random_2
```

### Shuffle the order of elements of a tensor

```python
not_shuffled = tf.constant([[10,7],
                            [2,3],
                            [4,5]])

#not_shuffled

#shuffle our non_shuffled tensor
tf.random.set_seed(5)  # global random seed
tf.random.shuffle(not_shuffled, seed=12) #operational random seed
```

> Rule 4: if both of type of seeds are set than the tensors with be shuffled in the same order

### Getting information from your tensors (tensor attributes)

Tensor attributes are:
- Shape
- Rank
- Axis or dim
- Size


```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape =(2,3,4,5))
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
2*3*4*5
```

```python
# Get various attributes of our tensor
print("Data type of every element:= ", rank_4_tensor.dtype)
print("Number of dim (rank):= ",rank_4_tensor.ndim)
print("Shape of our tensor:= ",rank_4_tensor.shape)
print("Elements along the 0 axis:=", rank_4_tensor[0])
print("Elements along the last axis:=", rank_4_tensor[-1])
print("The size of the tensor:= ",tf.size(rank_4_tensor))
print("The size of the tensor:= ",tf.size(rank_4_tensor).numpy())
```

### Change the datatype of a tensor

```python
# Create a new tensor with default data type
B = tf.constant([1.7,7.4])
B, B.dtype
```

```python
# Get the absolute values
tf.abs(D)
```

### Find the index position of maximum and minimum

```python
# Create a new tensor for finding the min and max index
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the index of the max element
tf.argmax(F)
```

```python
# Find the max value (element) of our tensor
F[tf.argmax(F)], tf.reduce_max(F)
```

```python
# check for equality between
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the index of min element
tf.argmin(F)
```

```python
# Find the min value (element) of our tensor
F[tf.argmin(F)], tf.reduce_min(F)
```
